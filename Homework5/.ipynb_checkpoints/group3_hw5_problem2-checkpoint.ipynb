First we import our numerical workhorses and our data set. 

```python
import numpy as np
import pandas as pd
import scipy.optimize
import numba
import matplotlib.pyplot as plt
import seaborn as sns
import emcee
import corner

%matplotlib inline

# import Justin's settings
rc = {'lines.linewidth': 2,
      'axes.labelsize': 18,
      'axes.titlesize': 18,
      'axes.facecolor': 'DFDFE5'}
sns.set_context('notebook', rc=rc)
sns.set_style('darkgrid', rc=rc)

df = pd.read_csv('./data/fret_binding_curve.csv', comment = '#')
```

We want to now compute the background-subtracted fluorescence and the FRET efficency. Where FRET efficency is given by

\begin{align}
e = 1 - \frac {F}{F_0}
\end{align}

Where F is the measured fluorescence signal and $F_0$ is the fluorescence signal when there is no b protein.

```python
# to compute the background subtracted fluorescence 
# we will just add a column called background-subtracted fluorescence to the dataframe

df['background-subtracted fluorescence'] = df['fluorescence'] - df['buffer']


# the fluorescence efficency is 1 - F/F_0
# where F_0 is the fluorescence when there is no b protein
# and F is the measured fluorescence signal
# we will create another column for this

df['fluorescence efficency'] = 1 - df['background-subtracted fluorescence'] / 437000.000

print(df)
```

# Part b

We want to perform a regression on $K_d$ with outliers. We first need to define our log posterior.

```python
# now we want to perform a regression to find K_d with an outlier detector scheme

# first we define a log posterior
# our log posterior is 


def f_efficiency(K_d, alpha, c_a_0, c_b_0):
    '''
    define efficiency
    '''
    
    #define what is in the sqrt of the denominator
    denom_sqrt = (K_d + c_a_0 + c_b_0)**2 - 4 * c_a_0 * c_b_0
    
    return 2 * alpha * c_b_0 / (K_d + c_a_0 + c_b_0 + np.sqrt(denom_sqrt))
                                
def resid(K_d, alpha, c_a_0, c_b_0, e):
    '''
    Residual for theorhetical spindle model.
    '''
    return e - f_efficiency(K_d, alpha, c_a_0, c_b_0)

def log_post(p, c_a_0, c_b_0, e):
    '''
    perform regressions to find K_d
    '''    
    # unpack parameters
    K_d, alpha = p
    
    # set boundaries
    if (alpha <= 0) or (alpha >= 1) or (K_d > 0):
        return -np.inf
    
    # define the log posterior
    return - len(c_a_0) / 2 * np.log(np.sum(resid(K_d, alpha, c_a_0, c_b_0, e)**2))
```

Now we can perform MCMC to retrieve our parameter values

```python
# define our arguments
args = (df['a conc (nM)'], df['b conc (nM)'], df['fluorescence efficency'])

# number of parameters we are going to sample
n_dim = 2

# wants to know how many walkers we are going to use
n_walkers = 50

# author of package says have at least twice as many walkers has you do parameters

# if walkers walk around long enough they will sample out of posterior distribution
# if you start at bad poitn don't want to count that cause not sampling appropiately
# this is how many steps you're going to burn
n_burn = 500

# specifiy how many steps you're going to take
n_steps = 5000

# seed random number generator for reproducibility
np.random.seed(42)

# we already defined our posteriors
# we need to tell the walkers where to start
# p0[i,j] is the starting point of walker i for parameter j
p0 = np.empty((n_walkers, n_dim))
p0[:,0] = np.random.exponential(0.1, n_walkers)
p0[:,1] = np.random.uniform(0, 1, n_walkers)
# r is column 0 p is column 1
# we are saying for p call random numbers from interval (0, 1) and call n_walkers
# now we have wehre the walkers start

# we have to instantiate our sampler
sampler = emcee.EnsembleSampler(n_walkers, n_dim, log_post, args = args)

# we need to burn the sampler in
# pos is the position, it's the one we care about
pos, prob, state = sampler.run_mcmc(p0, n_burn, storechain = False)
# it throws away everything and just returns the final position

# sample
_ = sampler.run_mcmc(pos, n_steps)

# parameter set
max_ind = np.argmax(sampler.flatlnprobability)

# Pull out values.
K_d_MAP, alpha_MAP = sampler.flatchain[max_ind,:]

# Print the results
print("""
Most probable parameter values:
K_d:  {0:.3f}
alpha: {1:.3f}
""".format(K_d_MAP, alpha_MAP))
```

Now we can plot a corner plot.

```python
fig = corner.corner(sampler.flatchain, labels=['K_d', 'alpha'], bins=100)
```

Now we want to preform a regression of $K_d$ without outliers. We do this by using a performing a cauchy distribution. We define a new log posterior.

```python
def f_efficiency(K_d, alpha, c_a_0, c_b_0):
    '''
    define efficiency
    '''
    
    #define what is in the sqrt of the denominator
    denom_sqrt = (K_d + c_a_0 + c_b_0)**2 - 4 * c_a_0 * c_b_0
    
    return 2 * alpha * c_b_0 / (K_d + c_a_0 + c_b_0 + np.sqrt(denom_sqrt))
                                
                           
def resid(K_d, alpha, beta, c_a_0, c_b_0, e):
    '''
    Residual for theorhetical spindle model.
    '''
    
    return (e - f_efficiency(K_d, alpha, c_a_0, c_b_0)) / beta

def log_post_cauchy(p, c_a_0, c_b_0, e):
    '''
    perform regressions to find K_d
    '''
    
    K_d, alpha, beta = p

    if (alpha <= 0) or (alpha > 1) or (K_d <= 0) or (beta <= 0):
        return -np.inf
    
    # define the log posterior
    return - (len(c_a_0) + 1) * np.log(beta) - np.sum(np.log(1 + resid(K_d, alpha, beta, c_a_0, c_b_0, e)**2))
```

Now we can perform MCMC.

```python
# define our arguments
args = (df['a conc (nM)'], df['b conc (nM)'], df['fluorescence efficency'])

# number of parameters we are going to sample
n_dim = 3

# wants to know how many walkers we are going to use
n_walkers = 50

# author of package says have at least twice as many walkers has you do parameters

# if walkers walk around long enough they will sample out of posterior distribution
# if you start at bad poitn don't want to count that cause not sampling appropiately
# this is how many steps you're going to burn
n_burn = 500

# specifiy how many steps you're going to take
n_steps = 5000

# seed random number generator for reproducibility
np.random.seed(42)

# we already defined our posteriors
# we need to tell the walkers where to start
# p0[i,j] is the starting point of walker i for parameter j
p0 = np.empty((n_walkers, n_dim))
p0[:,0] = np.random.exponential(0.1, n_walkers)
p0[:,1] = np.random.uniform(0, 1, n_walkers)
p0[:,2] = np.random.exponential(0.1, n_walkers)
# r is column 0 p is column 1
# we are saying for p call random numbers from interval (0, 1) and call n_walkers
# now we have wehre the walkers start

# we have to instantiate our sampler
sampler = emcee.EnsembleSampler(n_walkers, n_dim, log_post_cauchy, args = args)

# we need to burn the sampler in
# pos is the position, it's the one we care about
pos, prob, state = sampler.run_mcmc(p0, n_burn, storechain = False)
# it throws away everything and just returns the final position

# sample
_ = sampler.run_mcmc(pos, n_steps)

# parameter set
max_ind = np.argmax(sampler.flatlnprobability)

# Pull out values.
K_d_MAP, alpha_MAP, beta_MAP = sampler.flatchain[max_ind,:]

# Print the results
print("""
Most probable parameter values:
K_d:  {0:.3f}
alpha: {1:.3f}
beta: {2:.3f}
""".format(K_d_MAP, alpha_MAP, beta_MAP))

fig = corner.corner(sampler.flatchain, labels=['K_d', 'alpha', 'beta'], bins=100)
```
