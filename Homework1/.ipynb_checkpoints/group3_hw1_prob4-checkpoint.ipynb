```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#in-line magic function
%matplotlib inline

# JB's favorite Seaborn settings for notebooks
rc={'lines.linewidth': 2, 'axes.labelsize': 18, 'axes.titlesize': 18,
    'axes.facecolor': 'DFDFE5'}
sns.set_context('paper', rc=rc)
```

**Elena's work**

I was interested in seeing whether beak length varies within the same species across different islands. Significant variation across islands could indicate early stages of differentiation into separate subspecies or even new species altogether.

First, I read the data from the .TAB file into a dataframe. 

```python
df = pd.read_csv('./data/morphLack.TAB', delimiter='\t')
```

I then sliced out only those entries belonging to Geospiza fuliginosa from four different islands, and concatenated into a new dataframe. I plotted the upper beak length vs. island ID as a strip-plot because there were not an excessive number of data points. 

```python
ful_df = df[df['TaxonOrig'] == 'Geospiza fuliginosa']

ful_Mrch_Bndl = ful_df[ful_df['IslandID'] == 'Mrch_Bndl'] 
ful_Isa_Alb = ful_df[ful_df['IslandID'] == 'Isa_Alb'] 
ful_Flor_Chrl = ful_df[ful_df['IslandID'] == 'Flor_Chrl']
ful_Balt_SS = ful_df[ful_df['IslandID'] == 'Balt_SS']

ful_islands = pd.concat([ful_Mrch_Bndl, ful_Isa_Alb, ful_Flor_Chrl, 
                         ful_Balt_SS])
                        
sns.stripplot(x=ful_islands['IslandID'], 
              y=ful_islands['UBeakL'], data=df, jitter=True, color='Blue',
              alpha=0.6) 
plt.ylabel('Upper Beak Length (mm)')
plt.xlabel('Island ID')  
plt.title('Geospiza fuliginosa')
```

At first glance, it looks like there is not a huge difference in the distribution of beak lengths across the four islands. However, the difference between Mrch_Bndl and Balt_SS (left-most and right-most) may be significant; further statistical analysis, such as by a t-test, would be required in order to support this intuition.

I then performed the same analysis for Geospiza scandens.

```python
sca_df = df[df['TaxonOrig'] == 'Geospiza scandens']

sca_Mrch_Bndl = sca_df[sca_df['IslandID'] == 'Mrch_Bndl'] 
sca_Isa_Alb = sca_df[sca_df['IslandID'] == 'Isa_Alb'] 
sca_Flor_Chrl = sca_df[sca_df['IslandID'] == 'Flor_Chrl']
sca_Balt_SS = sca_df[sca_df['IslandID'] == 'Balt_SS']

sca_islands = pd.concat([sca_Mrch_Bndl, sca_Isa_Alb, sca_Flor_Chrl, 
                         sca_Balt_SS])
 
sns.stripplot(x=sca_islands['IslandID'], 
              y=sca_islands['UBeakL'], data=df, jitter=True, color='Red',
              alpha=0.6)               
plt.ylabel('Upper Beak Length (mm)')
plt.xlabel('Island ID')             
plt.title('Geospiza scandens')
```

It looks like the distribution for Flor_Chrl is quite different from that of Balt_SS and Mrch_Bndl, and may also be significantly different from Isa_Alb. Again, statistical analysis by a series of pair-wise comparisons. Since we would want to compare more than one pair, the Tukey HSD test might be appropriate. 

The differences in beak length distribution across the four islands are not the same for both species that we looked at. I would hypothesize that the differences (if they are statistically significant) are not necessarily due to features of the islands themselves (which might bias all species towards longer or shorter beaks) but may be due to different community compositions and hence different competitive interactions between species in procuring food.
