First we import `numpy`, `seaborn`, `pandas`, and `matplotlib`. Then we need to read our data into a `DataFrame`, we use the `read_csv delimiter='\t'` arguement to do so. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import beeswarm as bs
import seaborn as sns
sns.set()

#in-line magic function
%matplotlib inline

# This enables high resolution PNGs. SVG is preferred, but has problems
# rendering vertical and horizontal lines
%config InlineBackend.figure_formats = {'png', 'retina'}

# JB's favorite Seaborn settings for notebooks
rc={'lines.linewidth': 2, 'axes.labelsize': 18, 'axes.titlesize': 18,
    'axes.facecolor': 'DFDFE5'}
sns.set_context('paper', rc=rc)


plt.close('all')

df = pd.read_csv('morphLack.csv', comment = '#', delimiter='\t')
```

Now we can splice this data. We can look at wing lenght first.

```python
print(df['WingL'])
```

We can get rid of the NaN values using the `pandas` `dropna()` arguement.

```python
print(df['WingL'].dropna())
```

We can compare the wing lengths of the finches observe on each island. 

```python
mean_Wing_len = df.groupby('IslandID')['WingL'].mean()
std_Wing_len = df.groupby('IslandID')['WingL'].std()

# We can plot a bar graph
mean_Wing_len.plot(kind='bar', yerr=std_Wing_len)

# So that the x axis labels do not over lap we can rotate them 90 degrees
plt.xticks(rotation = 90)
plt.ylabel('Wing Length')
```

But this doesn't tell us much and doesn't look that good. What we can do is we can represent our plot using `seaborn` argument `boxplot()`. If our data is tidy, it should give us a very clean and easy to understand representation of out data. 

```python
# Data is tidy, so we should be able to represent our data as a box plot using seaborn
sns.boxplot(x = 'IslandID', y = df['WingL'].dropna(), data = df, width = 0.5)

# So that the x axis labels do not over lap we can rotate them 90 degrees
plt.xticks(rotation = 90)

# We don't want the column names as the axes labels so we can change them
plt.ylabel('Wing Length')
plt.xlabel('Island')
```

Now we can also compare wing length across species.

```python
# Data is tidy, so we should be able to represent our data as a box plot using seaborn
sns.boxplot(x = 'SpeciesL69', y = df['WingL'].dropna(), data = df, width = 0.5)

# So that the x axis labels do not over lap we can rotate them 90 degrees
plt.xticks(rotation = 90)

# We don't want the column names as the axes labels so we can change them
plt.ylabel('Wing Length')
plt.xlabel('Species')
```

So now we have an idea of how the wing length of finches compare by island and by species. Let's take a look and compare the wing length between species on the same island. 

```python
# splice just Wing length and Island ID
wing_len_island = df[['WingL','IslandID']].dropna()

# Data is tidy, so we should be able to represent our data as a box plot using seaborn
sns.boxplot(x = 'SpeciesL69', y = df['WingL'].dropna(), data = df, width = 0.5)

# So that the x axis labels do not over lap we can rotate them 90 degrees
plt.xticks(rotation = 90)

# We don't want the column names as the axes labels so we can change them
plt.ylabel('Wing Length')
plt.xlabel('Species')
```
